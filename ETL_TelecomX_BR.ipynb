# üìä Projeto Churn de Clientes ‚Äì Telecom X

## 1. Introdu√ß√£o

O objetivo deste projeto √© analisar o comportamento dos clientes da empresa **Telecom X** para compreender os fatores que influenciam a **evas√£o de clientes (churn)**. Atrav√©s de um processo de **ETL (Extra√ß√£o, Transforma√ß√£o e Carga)** e **an√°lise explorat√≥ria de dados (EDA)**, buscamos obter insights relevantes que auxiliem na cria√ß√£o de estrat√©gias para reten√ß√£o de clientes.

---

## 2. Limpeza e Tratamento de Dados

A primeira etapa envolveu a **extra√ß√£o dos dados diretamente de uma API (arquivo JSON)** hospedado no GitHub, convertendo-os para um **DataFrame do Pandas** para facilitar sua manipula√ß√£o.

Durante o processo de tratamento:

- Foram identificados e tratados valores **nulos e inconsistentes**, especialmente na coluna `Churn`, onde registros vazios foram tratados de duas formas:
  - Dataset limpo com apenas registros v√°lidos de churn (`df_churn_limpos`).
  - Dataset completo com valores vazios preenchidos como `"Unknown"` (`df_churn_unknown`), mantendo integridade para an√°lises n√£o supervisionadas.

- Vari√°veis categ√≥ricas bin√°rias com valores `"Yes"` e `"No"` foram convertidas para **1 e 0**, facilitando an√°lises quantitativas.
- Foi criada a coluna **`Daily_Charges`** com base no valor m√©dio mensal dividido por 30,42 dias, proporcionando uma vis√£o mais granular dos gastos di√°rios dos clientes.
- Dados duplicados e outliers foram verificados, e o tipo de dados das vari√°veis foi padronizado.

---

## 3. An√°lise Explorat√≥ria de Dados (EDA)

A an√°lise explorat√≥ria visou entender o comportamento geral dos clientes e sua rela√ß√£o com a evas√£o, por meio de **gr√°ficos, medidas estat√≠sticas e visualiza√ß√µes multivariadas**.

### 3.1. Perfil dos Clientes que Evadem

- A taxa de churn foi **desbalanceada**, com uma **propor√ß√£o menor de clientes que evadiram**.
- Clientes com **baixo tempo de contrato (`tenure`)** apresentam maior propens√£o a churn.
- O churn √© mais alto entre:
  - Clientes **sem parceiro ou dependentes**.
  - Clientes com **fatura digital ativada (paperless billing)**.
  - Usu√°rios com **m√©todo de pagamento eletr√¥nico** (como cart√£o ou cobran√ßa autom√°tica).
  - Clientes com **servi√ßos de internet do tipo "Fiber optic"**.

- Contratos **mensais** apresentaram significativamente mais churn do que contratos anuais ou bianuais.

### 3.2. Correla√ß√µes com Churn

- O **tempo de perman√™ncia do cliente (`tenure`)** mostrou **forte correla√ß√£o negativa** com churn.
- **Total de cobran√ßas (`Charges.Total`)** tamb√©m teve correla√ß√£o negativa com churn, indicando que clientes com hist√≥rico mais longo e faturamento acumulado tendem a permanecer.
- O **valor da cobran√ßa mensal** por si s√≥ n√£o apresentou correla√ß√£o forte, indicando que o valor em si n√£o √© o principal fator isolado.

- O **heatmap de correla√ß√£o** mostrou que poucas vari√°veis num√©ricas t√™m alta correla√ß√£o entre si, refor√ßando que churn depende de **v√°rias pequenas contribui√ß√µes combinadas**.

---

## 4. Conclus√µes e Insights

- **Clientes com contratos mensais**, especialmente com **pagamento autom√°tico e fatura digital**, t√™m maior tend√™ncia √† evas√£o.
- A **longevidade** √© um fator crucial: quanto maior o tempo de relacionamento com a empresa, **menor a chance de churn**.
- Usu√°rios de **internet via fibra** t√™m maior probabilidade de churn, sugerindo insatisfa√ß√£o com o servi√ßo ou concorr√™ncia mais forte nesse segmento.
- **Servi√ßos adicionais** como **seguran√ßa online, suporte t√©cnico e backup** est√£o correlacionados com **menor churn**, indicando que **clientes mais engajados nos servi√ßos tendem a permanecer**.

---

## 5. Recomenda√ß√µes

Com base nos dados analisados, s√£o sugeridas as seguintes a√ß√µes:

1. **Foco na reten√ß√£o de clientes novos:** criar campanhas espec√≠ficas para clientes com menos de 6 meses de contrato.

2. **Incentivar contratos mais longos** com benef√≠cios e descontos para reduzir a taxa de churn de contratos mensais.

3. **Revisar a experi√™ncia dos clientes com internet de fibra**, investigando se h√° falhas t√©cnicas ou concorr√™ncia desleal.

4. **Promover o uso de servi√ßos adicionais** (como backup, seguran√ßa, suporte t√©cnico) com campanhas de ades√£o gratuita por um per√≠odo, buscando aumentar o engajamento e reten√ß√£o.
